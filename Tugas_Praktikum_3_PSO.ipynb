```python
import numpy as np
import matplotlib.pyplot as plt

# Mengatur seed agar hasil simulasi tetap sama setiap kali dijalankan (opsional)
np.random.seed(42)
```

```python
# 1. Parameter Simulasi
n_produksi = 1000  # Jumlah barang yang diproduksi per hari
p_cacat = 0.02     # Probabilitas cacat (2%)
hari = 30          # Durasi simulasi 30 hari ke depan

# Melakukan simulasi jumlah barang cacat selama 30 hari
data_cacat = np.random.binomial(n_produksi, p_cacat, hari)

print("Hasil simulasi jumlah barang cacat per hari:")
print(data_cacat)
```

# Simulasi Distribusi Binomial

Pada cell ini, kita menjalankan inti dari simulasi stokastik menggunakan fungsi np.random.binomial. Berdasarkan output yang muncul, berikut adalah poin-poin penjelasannya:
## Input Parameter

Kode tersebut menggunakan tiga parameter utama yang menentukan hasil simulasi:
* $n$ (n_produksi = 1000): Jumlah percobaan dalam satu hari (total barang yang diproduksi).
* $p$ (p_cacat = 0.02): Peluang terjadinya "kejadian sukses" (dalam hal ini, barang yang cacat) untuk setiap unit.
* Size (hari = 30): Jumlah pengulangan simulasi, yaitu selama 30 hari.2.

## Memahami Output Array

Output yang kamu lihat di bawah kode tersebut adalah sebuah Array (kumpulan data) yang berisi 30 angka.
* Setiap angka mewakili satu hari: Angka pertama 18 adalah jumlah barang cacat di hari ke-1, angka kedua 28 di hari ke-2, dan seterusnya hingga hari ke-30.
* Sifat Stokastik: Perhatikan bahwa meskipun rata-rata teoritisnya adalah 20 unit ($1000 \times 0.02$), hasil harian bervariasi (ada yang serendah 12 dan setinggi 29). Inilah yang disebut dengan variasi acak dalam simulasi.
## Logika Distribusi Binomial

Fungsi ini bekerja seperti melempar koin yang tidak seimbang sebanyak 1000 kali setiap harinya.
* Karena probabilitas cacatnya kecil (2%), maka angka-angka yang muncul akan cenderung berkumpul di sekitar angka 20.
* Munculnya angka seperti 12 atau 29 menunjukkan adanya fluktuasi yang mungkin terjadi di dunia nyata, meskipun proses produksinya tetap sama.

```python
# 2. Membuat Histogram
plt.figure(figsize=(10, 6))
plt.hist(data_cacat, bins=range(min(data_cacat), max(data_cacat) + 2),
         color='skyblue', edgecolor='black', align='left')

plt.title('Histogram Hasil Simulasi Barang Cacat (30 Hari)')
plt.xlabel('Jumlah Barang Cacat per Hari')
plt.ylabel('Frekuensi Muncul')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

# Visualisasi Distribusi (Histogram)

Setelah mendapatkan data simulasi selama 30 hari, kita memvisualisasikannya ke dalam Histogram. Histogram ini berfungsi untuk melihat pola penyebaran dan frekuensi kejadian jumlah barang cacat.

## Cara Membaca Grafik:

* Sumbu X (Horizontal): Menunjukkan jumlah barang yang cacat dalam satu hari (rentang antara 12 hingga 29 unit).

* Sumbu Y (Vertical): Menunjukkan frekuensi atau berapa kali jumlah cacat tersebut muncul dalam periode 30 hari.

* Interpretasi Data: * Terlihat bahwa batang tertinggi berada di angka 16, yang berarti dalam 30 hari simulasi, kejadian dengan 16 barang cacat muncul paling sering (sebanyak 6 kali).

* Data cenderung berkumpul (terpusat) di area pertengahan (sekitar 16-21), yang menunjukkan bahwa proses produksi secara umum menghasilkan cacat di angka tersebut.

* Nilai ekstrem seperti 12 atau 29 memiliki frekuensi rendah, yang berarti kejadian tersebut jarang terjadi.

## Kesimpulan Visual:

Grafik ini mengonfirmasi bahwa meskipun prosesnya acak, hasilnya tetap mengikuti pola Distribusi Binomial di mana nilai-nilai yang dekat dengan nilai rata-rata memiliki peluang muncul yang lebih tinggi.

```python
# 3. Hitung Rata-rata dan Standar Deviasi
rata_rata_cacat = np.mean(data_cacat)
std_deviasi_cacat = np.std(data_cacat)

print(f"Rata-rata barang cacat per hari: {rata_rata_cacat:.2f}")
print(f"Standar deviasi barang cacat: {std_deviasi_cacat:.2f}")
```

# Analisis Statistik: Rata-rata dan Standar Deviasi
Setelah melakukan simulasi, kita merangkum data tersebut menggunakan dua indikator statistik utama untuk memahami performa produksi secara keseluruhan.Interpretasi Hasil:
## Rata-rata (Mean) - 19.10:

* Nilai ini menunjukkan bahwa secara rata-rata, terdapat sekitar 19 unit barang cacat setiap harinya.
* Hasil ini sangat mendekati nilai harapan matematis kita ($1000 \times 0.02 = 20$). Perbedaan kecil (19.10 vs 20) terjadi karena sifat acak dari simulasi dalam jangka pendek (hanya 30 hari).

## Standar Deviasi - 4.09:

* Standar deviasi mengukur seberapa jauh data bervariasi dari nilai rata-ratanya.
* Nilai 4.09 menunjukkan bahwa sebagian besar jumlah cacat harian akan berkisar di antara $19.10 \pm 4.09$ (yaitu antara 15 hingga 23 unit).
* Semakin kecil nilai ini, semakin konsisten (stabil) proses produksinya. Nilai 4.09 dalam konteks produksi 1000 unit dianggap relatif rendah, yang berarti penyimpangan yang terjadi tidak terlalu ekstrem.

## Kesimpulan:
Statistik ini memberikan kepastian bagi pihak manajemen bahwa meskipun jumlah cacat berubah-ubah setiap hari, angkanya tetap terkendali di sekitar nilai rata-rata yang dapat diprediksi.

```python
# 4. Analisis Hasil
# Kita bandingkan standar deviasi dengan nilai rata-rata
print("--- Analisis ---")
if std_deviasi_cacat / rata_rata_cacat < 0.25:
    print("Hasil: Jumlah barang cacat cenderung STABIL.")
    print("Penjelasan: Variasi harian relatif kecil dibandingkan dengan rata-ratanya.")
else:
    print("Hasil: Jumlah barang cacat cenderung BERFLUKTUASI.")
    print("Penjelasan: Terdapat perbedaan yang cukup signifikan antar hari produksi.")

# Menampilkan grafik garis untuk melihat tren harian
plt.figure(figsize=(10, 4))
plt.plot(data_cacat, marker='o', linestyle='-', color='red')
plt.title('Tren Harian Barang Cacat')
plt.xlabel('Hari Ke-')
plt.ylabel('Jumlah Cacat')
plt.xticks(range(hari))
plt.grid(True)
plt.show()
```

# Analisis Hasil: Stabilitas vs Fluktuasi
Bagian terakhir ini bertujuan untuk menyimpulkan kondisi operasional pabrik berdasarkan data simulasi selama 30 hari.

## Kesimpulan Status: STABIL
Berdasarkan perbandingan antara Standar Deviasi (4.09) dan Rata-rata (19.10), jumlah barang cacat harian dikategorikan cenderung STABIL. Hal ini dikarenakan variasi harian tidak melompat terlalu jauh dari angka rata-ratanya secara ekstrem jika dibandingkan dengan total produksi harian (1000 unit).

## Membaca Grafik Tren Harian (Line Chart):

* Fluktuasi Terkendali: Meskipun grafik terlihat naik-turun secara dinamis (zigzag), pergerakannya tetap berada dalam batas wajar distribusi binomial.

* Pola Acak: Tidak terlihat adanya tren naik terus-menerus atau turun terus-menerus. Ini menunjukkan bahwa cacat yang terjadi murni karena faktor keberuntungan/acak (stokastik), bukan karena adanya kerusakan sistematis pada mesin produksi.

* Puncak dan Lembah: Titik terendah (Hari ke-10) dan titik tertinggi (Hari ke-11) menunjukkan bahwa dalam operasional nyata, manajemen harus siap dengan skenario terburuk di mana cacat bisa melonjak tiba-tiba, namun kemungkinan besar akan kembali ke angka rata-rata di hari berikutnya.

## Analisis Akhir:
Proses produksi ini memiliki tingkat kendali kualitas yang baik. Fluktuasi yang terjadi masih dalam batas toleransi statistik, sehingga tidak memerlukan tindakan perbaikan darurat kecuali jika nilai rata-rata atau standar deviasinya meningkat secara signifikan di masa depan.
